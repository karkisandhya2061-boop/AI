Task 3: Missing Values & Data Cleaning

```python
print(hdi_2022_df.columns)
```

```python

print("Missing values:")
print(hdi_2022_df.isnull().sum())

hdi_2022_df['gross_inc_percap'] = (
    hdi_2022_df['gross_inc_percap']
    .replace("–", np.nan)
    .astype(float)
)

hdi_2022_df = hdi_2022_df.drop_duplicates()

hdi_2022_df = hdi_2022_df.dropna(subset=['hdi'])
```

Basic Statistics

```python


mean_hdi = hdi_2022_df['hdi'].mean()
median_hdi = hdi_2022_df['hdi'].median()
std_hdi = hdi_2022_df['hdi'].std()

print("Mean HDI:", mean_hdi)
print("Median HDI:", median_hdi)
print("Standard Deviation:", std_hdi)

print("\nCountry with Highest HDI:")
print(hdi_2022_df.loc[hdi_2022_df['hdi'].idxmax()])


print("\nCountry with Lowest HDI:")
print(hdi_2022_df.loc[hdi_2022_df['hdi'].idxmin()])
```

Filtering & Sorting

```python

high_hdi_df = hdi_2022_df[hdi_2022_df['hdi'] > 0.800]

high_hdi_sorted = high_hdi_df.sort_values(
    by='gross_inc_percap',
    ascending=False
)

high_hdi_sorted.head(10)
```

Add HDI Category Column

```python


def hdi_category(hdi):
    if hdi < 0.550:
        return "Low"
    elif hdi < 0.700:
        return "Medium"
    elif hdi < 0.800:
        return "High"
    else:
        return "Very High"

hdi_2022_df['HDI Category'] = hdi_2022_df['hdi'].apply(hdi_category)

hdi_2022_df['HDI Category'].value_counts()
```

Save Final Dataset

```python
hdi_2022_df.to_csv("HDI_category_added.csv", index=False)
```

Problem - 1B - HDI Visualization and Trend Analysis (2020 – 2022)

Data Extraction and Saving:

```python

hdi_trend_df = hdi_df.sort_values(by='year')

hdi_trend_df[['country', 'year', 'hdi']].head()
```

Data Cleaning

```python

essential_cols = ['hdi', 'country', 'year']
print("Missing values in essential columns:")
print(hdi_trend_df[essential_cols].isnull().sum())


hdi_trend_df = hdi_trend_df.dropna(subset=essential_cols)


before_duplicates = hdi_trend_df.shape[0]
hdi_trend_df = hdi_trend_df.drop_duplicates()
after_duplicates = hdi_trend_df.shape[0]

print(f"Duplicates removed: {before_duplicates - after_duplicates}")


numeric_cols = ['gross_inc_percap']

for col in numeric_cols:
    hdi_trend_df[col] = (
        hdi_trend_df[col]
        .replace("–", np.nan)
        .astype(float)
    )


hdi_trend_df['country'] = (
    hdi_trend_df['country']
    .str.strip()
    .str.title()
)


hdi_trend_df['year'] = hdi_trend_df['year'].astype(int)
```

Visualization Tasks:

A. Line Chart — HDI Trend (Country-Level):

```python
import matplotlib.pyplot as plt

selected_countries = ['Norway', 'Switzerland', 'Nepal', 'India', 'United States']


line_df = hdi_trend_df[
    (hdi_trend_df['country'].isin(selected_countries)) &
    (hdi_trend_df['year'].isin([2020, 2021, 2022]))
]


plt.figure(figsize=(10,6))

for country in selected_countries:
    country_data = line_df[line_df['country'] == country]
    plt.plot(country_data['year'], country_data['hdi'], marker='o', label=country)

plt.xlabel("Year")
plt.ylabel("HDI")
plt.title("HDI Trend (2020–2022) for Selected Countries")
plt.legend()
plt.grid(True)
plt.show()
```

**Interpretation**

The line graph represents the dynamics of Human Development Index (HDI) in five countries of choice in the 3-year timeframe of 2020-2022.

The high score of HDI of developed countries like Norway, Switzerland, and the United States has remained high (above 0.900) with minimal variations in the score of these countries on an annual basis.

These nations have a stable situation in the important development indicators such as life expectancy, education and income.

Conversely, the developing nations such as Nepal and India record a slow incremental trend, which demonstrates continuous enhancement in health, education and economic status.

The HDI score of Nepal and India is lower than 0.800 though it is experiencing a gradual improvement annually.

Notably, none of the countries in the graph has a significant decrease in HDI within this period, which shows relative stability and positive development trends.

The graphical comparison shows the unchanging divide within high and low-income nations, but also presents positive developments of emerging economies.**bold text**

Composite Development Score

```python
cols = ['life_expectancy', 'gross_inc_percap', 'hdi']
for col in cols:
    south_asia_df[col] = pd.to_numeric(south_asia_df[col], errors='coerce')

south_asia_df['Composite Score'] = (
    0.30 * south_asia_df['life_expectancy'] +
    0.30 * south_asia_df['gross_inc_percap']
)

composite_rank = south_asia_df.sort_values(
    by='Composite Score', ascending=False
)

composite_rank[['country', 'Composite Score', 'hdi']]
```

Outlier Detection (IQR Method)

```python
def detect_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return (series < lower) | (series > upper)

south_asia_df['HDI_Outlier'] = detect_outliers(south_asia_df['hdi'])
south_asia_df['GNI_Outlier'] = detect_outliers(south_asia_df['gross_inc_percap'])

south_asia_df[['country', 'HDI_Outlier', 'GNI_Outlier']]
```

```python

middle_east_countries = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan",
    "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
    "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"
]

# Use hdi_trend_df as the base DataFrame for filtering years 2020-2022
df_2020_22 = hdi_trend_df.loc[hdi_trend_df['year'].between(2020, 2022)].copy()

# Create south_asia_df for this problem context by filtering df_2020_22
# using the predefined list of south_asia_countries
south_asia_df = df_2020_22.loc[
    df_2020_22['country'].isin(south_asia_countries)
].copy()
south_asia_df['region'] = 'South Asia' # Assign the region explicitly for this subset

# Create middle_east_df by filtering df_2020_22 with the list of middle_east_countries
middle_east_df = df_2020_22.loc[
    df_2020_22['country'].isin(middle_east_countries)
].copy()
middle_east_df['region'] = 'Middle East' # Assign the region explicitly for this subset
```
